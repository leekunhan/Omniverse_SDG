```python
import omni.replicator.core as rep
import os

gpu_mats = rep.get.material(path_pattern = "/sticker_texture/GPU")
clip_mats = rep.get.material(path_pattern = "/sticker_texture/clip_label")
product_mats = rep.get.material(path_pattern = "/sticker_texture/product_sticker")
shipping_mats = rep.get.material(path_pattern = "/sticker_texture/shipping_sticker")
upd_mats = rep.get.material(path_pattern = "/sticker_texture/UPD_label")
upgrade_mats = rep.get.material(path_pattern = "/sticker_texture/Upgrade_label")
sn_mats = rep.get.material(path_pattern = "/sticker_texture/SN")
string_mats = rep.get.material(path_pattern = "/sticker_texture/sticker_string")

def clip_label_change():
    #設定哪一個plane要被轉換
    clip_label = rep.get.prims(path_pattern = "/box_with_sticker_01/clip_label/Plane")
    with clip_label:
        rep.modify.pose(position=rep.distribution.uniform((-0.03184, 0, -0.06), (0.11, 0, 0.03)),
                        rotation = rep.distribution.uniform((-90, 178, 0),(-90, 182, 0)))
        rep.modify.visibility(rep.distribution.choice([True,False,True]))
        rep.randomizer.materials(clip_mats)
rep.randomizer.register(clip_label_change)

def gpu_change():
    gpu = rep.get.prims(path_pattern = "/box_with_sticker_01/gpu/Plane")
    with gpu:
        rep.modify.pose(position=(0, 0, 0),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
        rep.modify.visibility(rep.distribution.choice([True,False,True]))
        rep.randomizer.materials(gpu_mats)
rep.randomizer.register(gpu_change)

def upd_change():
    upd = rep.get.prims(path_pattern = "/box_with_sticker_01/upd_label/Plane")
    with upd:
        rep.modify.pose(position=rep.distribution.uniform((-0.03, -0.01145, -0.005), (0.03, -0.01145, 0.06)),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
        rep.modify.visibility(rep.distribution.choice([True,False,True,True]))
        rep.randomizer.materials(upd_mats)
rep.randomizer.register(upd_change)

def upgrade_change():
    upgrade = rep.get.prims(path_pattern = "/box_with_sticker_01/upgrade_label/Plane")
    with upgrade:
        rep.modify.pose(position=rep.distribution.uniform((-0.07, -0.005, -0.16), (0, -0.005, 0.1)),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
        rep.modify.visibility(rep.distribution.choice([True,False,True,True]))
        rep.randomizer.materials(upgrade_mats)
rep.randomizer.register(upgrade_change)

def sn_change():
    sn = rep.get.prims(path_pattern = "/box_with_sticker_01/sn/Plane")
    with sn:
        rep.modify.pose(position=rep.distribution.uniform((-0.08, -0.01, -0.21), (0.08, -0.01, 0.09)),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
        #rep.modify.visibility(rep.distribution.choice([True,False,True,True]))
        rep.randomizer.materials(sn_mats)
rep.randomizer.register(sn_change)

def product_change():
    product = rep.get.prims(path_pattern = "/box_with_sticker_01/product_sticker/Plane")
    with product:
        rep.randomizer.materials(product_mats)
        rep.modify.pose(
            rotation = rep.distribution.uniform((90, -4, 90),(90, 4, 90))
        )
rep.randomizer.register(product_change)

def shipping_change():
    shipping = rep.get.prims(path_pattern = "/box_with_sticker_01/shipping_sticker/Plane")
    with shipping:
        rep.randomizer.materials(shipping_mats)
        rep.modify.pose(
            rotation = rep.distribution.uniform((90, -4, 90),(90, 4, 90))
        )
rep.randomizer.register(shipping_change)

def string_change():
    string = rep.get.prims(path_pattern = "/box_with_sticker_01/sticker_string/Plane")
    with string:
        rep.modify.visibility(rep.distribution.choice([True,False,True,True]))
        rep.randomizer.materials(string_mats)
        rep.modify.pose(scale = rep.distribution.uniform((0.5, 0.4, 1),(1.1, 0.6, 1)),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
rep.randomizer.register(string_change)

def string_change_01():
    string = rep.get.prims(path_pattern = "/box_with_sticker_01/sticker_string_01/Plane")
    with string:
        rep.modify.visibility(rep.distribution.choice([True,False,True,True]))
        rep.randomizer.materials(string_mats)
        rep.modify.pose(scale = rep.distribution.uniform((0.5, 0.4, 1),(1.1, 0.45, 1)),
                        position=rep.distribution.uniform((-0.43014, -0.001, -0.09118), (0.33586, -0.001, 0.00802)),
                        rotation = rep.distribution.uniform((90, -2, 0),(90, 2, 0)))
rep.randomizer.register(string_change_01)

def camera_far_control():
    camera = rep.get.prims(path_pattern = '/Camera_far')
    with camera:
        rep.modify.pose(
            position = rep.distribution.uniform(lower = (-450, 607, 11),upper =  (-435, 620, 17)),
            rotation = rep.distribution.uniform((84, -2, -189),(100.5, 2, -173))
        )
rep.randomizer.register(camera_far_control)

def camera_close_control():
    camera = rep.get.prims(path_pattern = '/Camera_close')
    with camera:
        rep.modify.pose(
            position = (-448.56211, 588.14709, 16.33158),
            rotation = rep.distribution.uniform((85.25171, -2, 173.95619),(94.45299, 2, 183.49875))
        )
rep.randomizer.register(camera_close_control)


with rep.trigger.on_frame(max_execs = 500, interval =50):
    rep.randomizer.clip_label_change()
    rep.randomizer.gpu_change()
    rep.randomizer.upd_change()
    rep.randomizer.upgrade_change()
    rep.randomizer.sn_change()
    rep.randomizer.shipping_change()
    rep.randomizer.product_change()
    rep.randomizer.string_change()
    rep.randomizer.string_change_01()
    rep.randomizer.camera_far_control()
    rep.randomizer.camera_close_control()
rep.orchestrator.preview()
```
